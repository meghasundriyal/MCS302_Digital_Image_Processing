```python
from PIL import Image
import math 
import numpy as np
```

```python
name="SampleImages/rose_gray.jpg"
```

```python
#opening unedited image
img = Image.open(name) 
img
```

```python
#calculate width and height of the image
width, height = img.size
```

```python
#creating image mask
imageMask = Image.open(name) 
width, height = imageMask.size
wmin, wmax = width/8, width/1.5
hmin, hmax = height/8, height/1.2

pixelsOfMask = imageMask.load()  

for w in range(width):
    for h in range(height):
        if(w > wmin and w < wmax and h > hmin and h < hmax):
            pixelsOfMask[w,h] = 255
        else:
            pixelsOfMask[w,h] = 0

imageMask.save("SampleImages/rose_gray_mask.jpg")
imageMask
```

```python
def booleanOr(m,p):
    new_pixel = (m | p)
    return round(new_pixel)
```

## performing boolean OR with imageMask

```python
img = Image.open(name)
#load pixel data
pixels = img.load()  

for w in range(width):
    for h in range(height):
        p = img.getpixel((w,h))
        m = imageMask.getpixel((w,h))
        pixels[w,h] = (booleanOr(m,p))

#save altered image        
# img.save("SampleImages/boolean_and.jpg")   
img
```

## performing boolean AND with imageMask

```python
def booleanAnd(m,p):
    new_pixel = (m & p)
    return round(new_pixel)
```

```python
img = Image.open(name)
#load pixel data
pixels = img.load()  

for w in range(width):
    for h in range(height):
        p = img.getpixel((w,h))
        m = imageMask.getpixel((w,h))
        pixels[w,h] = (booleanAnd(m,p))

#save altered image        
# img.save("SampleImages/boolean_and.jpg")   
img
```
