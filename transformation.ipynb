```python
from PIL import Image
import math 
import numpy as np
```

```python
#function for log transformation
def logTransform(c,pixel):
    new_pixel = c*math.log10(float(pixel+1))
    return round(new_pixel)
```

```python
# def main():
#open the image 
#calculate width and height of the image
img = Image.open("SampleImages/rose_gray.jpg")   
width, height = img.size

#load pixel data
pixels = img.load()  

#The scaling constant c is chosen so that the maximum output value is 255
c  = 255/(math.log10(1+ np.amax(img)))
print(c," C")
#going to each pixel of the image and adding a value to it
for w in range(width):
    for h in range(height):
        #retrieve the pixel value
        p = img.getpixel((w,h))
        #log tranform each pixel 
#         print(pixels[w,h]," before")
        pixels[w,h] = (logTransform(c,p))
    
#         print(pixels[w,h]," after")
        
#save altered image        
#     img.save("SampleImages/log_transformed.jpg")
#     print("here")
#     img.show()
img
```

```python
#function for log transformation
def powerLawTransform(c,pixel,gamma):
    new_pixel = c*pow(pixel,(1/gamma))
    return round(new_pixel)
```

```python
#gamma transformation
img = Image.open("SampleImages/rose_gray.jpg")   
width, height = img.size

#load pixel data
pixels = img.load()  


gamma = float(input("Enter a value for gamma : "))
#The scaling constant c is chosen so that the maximum output value is 255
pixel=np.amax(img)

c = 255/pow(pixel,(1/gamma))
print(c," C")
#going to each pixel of the image and adding a value to it
for w in range(width):
    for h in range(height):
        #retrieve the pixel value
        p = img.getpixel((w,h))
       
        pixels[w,h] = int(powerLawTransform(c,p,gamma))

img
```
