```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy.misc import imsave
import os
```

```python
open_location = "SampleImages/averaging_rgb/"
save_location = "SampleImages/averaging_gray/"
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
def convertToGrayscale(filename):
    img = mpimg.imread(open_location+filename)     
    gray = rgb2gray(img)  
    imsave(save_location+filename,gray)
```

```python
dirname = open_location
for filename in os.listdir(dirname):
    convertToGrayscale(filename)
```
