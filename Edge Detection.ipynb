```python
from PIL import Image
import math 
import numpy as np
```

```python
name="SampleImages/rose_gray.jpg"
```

```python
#opening unedited image
img = Image.open(name) 
img
```

```python
#calculate width and height of the image
width, height = img.size
pixels=img.load()
```

## Sobel Operator : 

```python
rows, cols = (3, 3) 
kernal_gx=[[-1,-2,-1],[0,0,0],[1,2,1]]
kernal_gy=[[-1,0,1],[-2,0,2],[-1,0,1]]
```

```python
edgeImage = Image.open(name)
edgePixels = edgeImage.load() 
widthEdgeImage , heightEdgeImage = edgeImage.size
```

```python
for w in range(1,width-1):
    for h in range(1,height-1):
        x=w-1
        y=h-1
        gx=0
        gy=0
        for r in range(rows):
            for c in range(cols):
                gx+= kernal_gx[r][c]*pixels[x,y]
                gy+= kernal_gy[r][c]*pixels[x,y]
                x+=1
            y+=1
            x=w-1
        
        edgePixels[w,h]=gx 
print("GX")
edgeImage        
```

```python
for w in range(1,width-1):
    for h in range(1,height-1):
        x=w-1
        y=h-1
        gx=0
        gy=0
        for r in range(rows):
            for c in range(cols):
                gx+= kernal_gx[r][c]*pixels[x,y]
                gy+= kernal_gy[r][c]*pixels[x,y]
                x+=1
            y+=1
            x=w-1
        
        edgePixels[w,h]=gy         
print("GY")

edgeImage
```

```python
for w in range(1,width-1):
    for h in range(1,height-1):
        x=w-1
        y=h-1
        gx=0
        gy=0
        for r in range(rows):
            for c in range(cols):
                gx+= kernal_gx[r][c]*pixels[x,y]
                gy+= kernal_gy[r][c]*pixels[x,y]
                x+=1
            y+=1
            x=w-1
        
        edgePixels[w,h]=gx+gy  
print("Gx+GY")
edgeImage
```
