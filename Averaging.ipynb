```python
from PIL import Image
import math 
import numpy as np
import os
```

# Average files in this directory

```python
dirname = "SampleImages/averaging_gray/"
```

```python
def averageImages(directory_name):
    directory_list = os.listdir(dirname)
    name = directory_list[0]
    avg_img = Image.open(directory_name+name)
    pixels_avg=avg_img.load()
    width_of_avg,height_of_avg=avg_img.size
    
    for w in range(width_of_avg):
        for h in range(height_of_avg):
            pixels_avg[w,h]=0
            
    image_matrix = np.empty((width_of_avg,height_of_avg),np.int64)
    
    length = len(directory_list)
    for filename in directory_list:
        img=Image.open(directory_name+filename)
        width,height=img.size
        if  width != width_of_avg or height != height_of_avg :
            
            return -1
        pixels=img.load()
        for w in range(width):
            for h in range(height):
                image_matrix[w][h] += pixels[w,h]
                
    for w in range(width_of_avg):
        for h in range(height_of_avg):
            pixels_avg[w,h] = (int(image_matrix[w][h]/length))
    return avg_img
```

```python
avg_img = averageImages(dirname)
if  avg_img == 0 :
    print("Incompatible Images")
else:
    avg_img
```
