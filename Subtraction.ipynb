```python
from PIL import Image
import math 
import numpy as np
```

```python
location="SampleImages/"
name1="rose_gray.jpg"

name2="log_transformed.jpg"
```

```python
def subtract(image_name_1,image_name_2):
    img1=Image.open(location+image_name_1)
    img2=Image.open(location+image_name_2)
    width1,height1=img1.size
    width2,height2=img2.size

    
    if width1 != width2  or height1 != height2:
        return -1
    pixels1 = img1.load()
    pixels2 = img2.load()
    for w in range(width1):
        for h in range(height1):
            pixels1[w,h] = (abs(pixels1[w,h]-pixels2[w,h]))  

    return img1
```

```python
difference=subtract(name1,name2)
```
